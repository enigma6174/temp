## Univariate Plotting

```python
import pandas as pd
reviews = pd.read_csv('winemag-data-130k-v2.csv', index_col=0)
reviews.head()
```

```python
# plot the wine-producing countries vs the number of labels of wine they produce (first 10 countries)
reviews['country'].value_counts().head(10).plot.bar()
```

```python
# what percentage of the total wine production is US made?
(reviews['country'].value_counts().head(10) / len(reviews)).plot.bar()
# US alone accounts for about 40% of total wine production
```

### Nominal Data
<ul>
    <li><font size="3"> that is purely categorical and has no sense of order to it; i.e. it cannot be ordered and can thus be compared only; best represented through
    bar charts </font></li>
    <li><font size="3"> eg. include sex (male/female), hair color, answers(yes/no/cant say) etc. </font></li>
    <li><font size="3"> Nominal Data is also called as categorical data </font></li>
</ul>
    
### Ordinal Data
<ul>
    <li><font size="3"> Categorical data with a sense of order is called as ordinal data; i.e. if the data can be categorized into discrete categories and can still be ordered
        it is called ordinal data </font></li>
    <li><font size="3"> eg. include t-shirt size (small/medium/large) etc. </font></li>
    <li><font size="3"> Ordinal data is best plotted using bar/line/area charts </font></li>
</ul>

```python
# display the counts of reviews to each point
reviews['points'].value_counts().sort_index().plot.bar()
```

```python
# display the counts of reviews to each point (as a line plot)
reviews['points'].value_counts().sort_index().plot.line()
```

```python
# plot the area chart of the count of the points
reviews['points'].value_counts().sort_index().plot.area()
```

### Interval Data
<font size = "3">
    <ul>
        <li>Interval data is ordinal data in which the periods (i.e. intervals) are spaced equally. They are always associated with continous data.</li>
        <li>Eg. of interval data include house price, temperature of a region, stock price etc.</li>
        <li>Line charts and histograms are best suited for interval data but in practice, histograms are preferred when plotting interval data.</li>
        <li>A histogram like a bar chart splits the data into bars but unlike a bar chart instead of each bar representing a single value, the bars of histogram represent 
a range of values (called an interval).</li>
        <li>The bars of histogram are called as 'bins'</li>
        <li>The biggest shortcoming of histograms is that they fail to convey useful information when the data is skewed as they tend to break up the space into evenly spaced intervals. (see second plot for example)</li>
    </ul>
</font>

```python
# plot a histogram of the prices of wines that are less than $200
reviews[reviews['price'] < 200]['price'].plot.hist()
#reviews.loc[reviews['price'] < 200]['price'].plot.hist() # another way of doing the above
```

```python
# an example of skewed data
reviews['price'].plot.hist()
```

```python
# plot a histogram of the points
reviews['points'].plot.hist()
```

### Scatter plot
<font size = "3">
    <ul>
        <li>A simple scatter plot maps each variable to a point in a 2D space.</li>
        <li>To make effective use of scatter plot, sometimes we need to <b>downsample</b> our data as native data scatter plots do not effectively treat points which map to the same value.</li>
        <li>If two or more points map to the same value, then all the subsequent points are plotted over the first one and if there are a large number of points mapped to the same value, we lose the structure of the scatter plot. </li>
        <li>Scatter plots thus work well with small datasets with large number of unique values</li>
</font>

```python
# display a scatter plot of the price (under $100) versus the points of 100 samples of wine
reviews[reviews['price'] < 100].sample(100).plot.scatter(x='price', y='points')
```

```python
# display a scatter plot of the points (less than 90) versus the countries of 100 samples of wine
# reviews[reviews['points'] < 90].sample(100).plot.scatter(x='points', y='country') # this is not allowed; look at the error message to figure out why
```

```python
# plotting without downsampling (price of bottle less than $100 versus the points)
reviews[reviews['price'] < 100].plot.scatter(x='price', y='points')
```

## Hexplot
<font size = "3">
    <ul>
        <li> A hex plot aggregates points in space into hexagons and then colors those hexagons based on the values in them.</li>
        <li> Hexplots work best with datasets that have large number of similar values and the intesity of the colour of each hexagon gives an estimate of the count of that points </li>
        <p><i><b>NOTE:</b> the x-axis label is missing due to a bug! </i></p>
    </ul>
</font>

```python
reviews[reviews['price'] < 100].plot.hexbin(x='price', y='points', gridsize=20)
```

```python
import pandas as pd
pd.set_option('max_columns', None)
df = pd.read_csv('CompleteDataset.csv', index_col=0)

import re
import numpy as np

footballers = df.copy()
footballers['Unit'] = df['Value'].str[-1]
footballers['Value (M)'] = np.where(footballers['Unit'] == '0', 0, 
                                    footballers['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
footballers['Value (M)'] = footballers['Value (M)'].astype(float)
footballers['Value (M)'] = np.where(footballers['Unit'] == 'M', 
                                    footballers['Value (M)'], 
                                    footballers['Value (M)']/1000)
footballers = footballers.assign(Value=footballers['Value (M)'],
                                 Position=footballers['Preferred Positions'].str.split().str[0])

footballers.head()
```

```python
import seaborn as sns

sns.lmplot(x='Value', y='Overall', hue='Position', data=footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])], fit_reg=False)
```

```python
sns.lmplot(x='Value', y='Overall', markers=['o', '+', '*'], hue='Position', data=footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])], fit_reg=False)
```

```python
f = (footballers.loc[footballers['Position'].isin(['ST', 'GK'])].loc[:,['Value', 'Overall', 'Aggression', 'Position']])
f = f[f['Overall'] >= 80]
f = f[f['Overall'] < 85]
f['Aggression'] = f['Aggression'].astype(float)

sns.boxplot(x='Overall', y='Aggression', hue='Position', data=f)
```

```python
f = (footballers.loc[:, ['Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control']].
     applymap(lambda v:int(v) if str.isdecimal(v) else np.nan).dropna()).corr()
sns.heatmap(f, annot=True)
```
